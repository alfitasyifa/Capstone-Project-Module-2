```python
# Memasukkan dataset 
df=pd.read_csv('Airbnb Listings Bangkok.csv')
# Menampilkan 5 baris teratas dan terakhir
display(df.head(),df.tail())
```

```python
df.drop(['Unnamed: 0'], axis = 1,inplace= True)
df
```

## Data Understanding and Cleaning


```python
# Mengetahui jumlah baris dan kolom
print(f'dataset memiliki jumlah baris dan kolom {df.shape}')
# Mengetahui informasi data
df.info()
```

```python
display(df.describe(), df.describe(include='object'))
```

```python
listDesc = []
for col in df.columns :
    listDesc.append([
        col,
        df[col].dtype,
        df[col].nunique(),
        df[col].drop_duplicates().sample(2).values
    ])

pd.DataFrame(listDesc, columns = ['Column Name', 'Data Type', 'Number Unique', 'Unique Sample'])
```

```python
# set as categorical data
df.room_type = df.room_type.astype('category')
df.neighbourhood = df.neighbourhood.astype('category')

# set as datetime data
df.last_review = pd.to_datetime(df.last_review)
```

```python
df.info()
```

**Missing Value**

```python
list = []
for col in df.columns :
    list.append( [col, df[col].isna().sum()])

tableNull = pd.DataFrame(columns=['Column Name', 'Null'],
                     data=list)
tableNull
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df[['name','host_name','last_review','reviews_per_month']].isna())
```

```python
df['name'].fillna('No Name', inplace= True)
df['host_name'].fillna('Unknown', inplace= True)
df['last_review'].fillna('0000-00-00', inplace= True)
df['reviews_per_month'].fillna('0', inplace= True)
```

```python
list = []
for col in df.columns :
    list.append( [col, df[col].isna().sum()])

tableNull_update = pd.DataFrame(columns=['Column Name', 'Null'],
                     data=list)
tableNull_update
```

# Data Analysis

```python
#jumlah berdasarkan tipe kamar
fig = sns.catplot(x = 'room_type', kind="count", data = df, order = df['room_type'].value_counts().index)
fig.set_xticklabels(rotation=45)
plt.show()
```

```python
df_neighbourhood=pd.crosstab(df['neighbourhood'],df['room_type'])
df_neighbourhood['Total']=df_neighbourhood.sum(axis=1)
df_neighbourhood.sort_values('Total',ascending=False).head()
```

```python
df_avg = df.groupby(by= "room_type")['price'].mean().reset_index()
df_avg = df_avg.sort_values("price", ascending= False)

plt.figure(figsize=(20, 10))
plt.bar(
    df_avg.room_type,
    df_avg.price
)
plt.title("Rata-Rata Harga per tipe kamar")
plt.xlabel("tipe kamar")
plt.ylabel("Rata-Rata Harga")
```

**Rekomendasi**
Memberikan penalti terhadap host yang memiliki ulasan <2 dari skala 5. Apabila penginapan memiliki ulasan buruk dalam kurun waktu 3 bulan, maka host tersebut terblokir dan tidak dapat didaftarkan kembali. Hal ini ditujukan agar host dapat meenjaga layaknya penginapan sesuai minat pelanggan. Selain itu, hal ini juga dapat mengurangi banyaknya penginapan yang kurang layak di wilayah yang banyak diminati wisatawan.
